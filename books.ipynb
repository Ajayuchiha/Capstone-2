filling null values

```python
dance_book.fillna({
    "book_title": "Unknown Title",
    "book_subtitle": "No Subtitle",
    "book_authors": "Unknown Author",
    "book_description": "No Description",
    "categories": "Uncategorized",
    "language": "Unknown",
    "imageLinks": "No Image",
    "ratingsCount": 0,
    "averageRating": 0.0,
    "country": "Unknown",
    "saleability": "NOT_FOR_SALE",
    "isEbook": False,
    "amount_listPrice": 0.0,
    "currencyCode_listPrice": "USD",
    "amount_retailPrice": 0.0,
    "currencyCode_retailPrice": "USD",
    "buyLink": "No Link",
    "year": "Unknown",
    "publisher": "Unknown Publisher"
}, inplace = True)
```

```python
dance_book.isnull().sum()
```

```python
dance_book['industryIdentifiers'] = dance_book['industryIdentifiers'].replace('[]','Unknown')
dance_book['imageLinks'] = dance_book['imageLinks'].replace('{}', 'Unknown')
```

```python
dance_book
```

```python
output_file = "dance_book_data.csv"
dance_book.to_csv(output_file, index = False)
```
